### 자동판매기 프로그램

- 자동 판매기를 시뮬레이션하는 프로그램을 작성
- 자동 판매기는 사용자로부터 투입한 돈과 물건 값을 입력 받는다
- 물건 값은 100원단위라고 가정 
- 프로그램은 잔돈을 계산하여 출력한다
- 자판기는 동전 500원, 100원짜리만가지고 있다고 가정

```python
money = int(input("투입한 돈"))
price = int(input("물건 값"))
change = money - price
print("거스름돈", change)
ch500 = change // 500
change = change % 500
ch100 = change // 100
print("500원 동전의 개수 : ", ch500)
print("100원 동전의 개수 : ", ch100)
```

### 원의 둘레, 넓이 , 구의 겉넓이, 부피구하기

```python
radius = int(input("반지름을 입력하시오"))
pi = 3.141592
print("원의 넓이", (radius ** 2) * pi)
print("원의 둘레", 2 * pi * radius)
print("구의 겉넓이", 4 * pi * (radius ** 2))
print("구의 부피", 4 / 3 * pi * (radius ** 3))
```

### 이중 루프를 이용해 신문 배달을 하는 프로그램을 작성하세요.
- 단, 아래에서 arrears 리스트는 신문 구독료가 미납된 세대에 대한 정보를 포함하고 있는데,
- 해당 세대에는 신문을 배달하지 않아야 합니다.

```python
apart = [[101, 102, 103, 104], [201, 202, 203, 204], [301, 302, 303, 304], [401, 402, 403, 404]]
arrears = [101, 203, 301, 404]

for floor in apart:
    for room in floor:
        if room in arrears:
            continue  
        print(f"신문 배달: {room}호")
```

```python
alibaba = "in a town of persia lived two brothers , sons of a poor man; one named cassim , the other alibaba . cassim , the elder"
vocab = {}
word_list = alibaba.split()
for word in word_list:
    if word not in vocab:
        vocab[word] = 1  
    else:
        vocab[word] += 1 

print(vocab)
```

```python
print(sorted(vocab.items(), key=operator.itemgetter(1), reverse = True))
```

```python
from collections import Counter
vocab = Counter(word_list)
print(vocab)
```

```python
#의미 없는 단어 제거
meaningless = ['a', 'the', 'in', 'of', ',']
for word in meaningless:
    if word in vocab:
        del vocab[word]
sorted(vocab.items(), key = operator.itemgetter(1), reverse = True)
```

```python
'''
동적 변수 : 화면이 생성된 후에 변하는 값을 저장하는 변수

사용 전에 먼저 동적 변수를 만들기
# 문자열 동적 변수
StringVar
# 정수형 동적 변수
Intvar
# 실수형 동적면수
DoubleVar
레이블 이름 = label(창이름, textvariable = 동적변수)
'''
```

```python
from tkinter import *
import random

n1, n2 = 0, 0

def new_prob():
    global n1, n2
    n1 = random.randint(2, 9)
    n2 = random.randint(1, 9)
    prob.set(f"{n1} x {n2} ?")
    answer_var.set('')  

def check_answer():
    answer_var.set(str(n1 * n2))

root = Tk()
root.title('구구단')

prob = StringVar()
q_lb1 = Label(root, textvariable=prob, width=15, font=('Arial', 20))
q_lb1.grid(row=0, column=0, padx=10, pady=10)

new_btn = Button(root, text='새로운 문제', width=12, command=new_prob)
new_btn.grid(row=0, column=1, padx=10, pady=10)

answer_var = StringVar()
ans_lb = Label(root, textvariable=answer_var, width=15, font=('Arial', 20))
ans_lb.grid(row=1, column=0, padx=10, pady=10)

ans_btn = Button(root, text='정답 확인', width=12, command=check_answer)
ans_btn.grid(row=1, column=1, padx=10, pady=10)

new_prob()

root.mainloop()
```
